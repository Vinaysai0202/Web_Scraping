```python
from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import wordcloud
```

```python
df = [review.strip() for review in df.review]
df = [review for review in  df if review]
df[0:10]
```

```python
text = " ".join(df)
text
```

```python
import string
no_punc_text = text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
text_tokens
len(text_tokens)
```

```python
text_tokens
```

```python
# removing stop_words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

my_stop_words = stopwords.words('english')
my_stop_words
len(my_stop_words)
```

```python
# removing stop words
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words ]
```

```python
# Normalize the data
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:20])
```

```python
# joining the words into single document (removing the tokenization)
doc =  ' '.join(lower_words)
doc
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()

lemmas = []

for token in doc.split():
    lemmas.append(Lemmatizer.lemmatize(token))
```

```python
# Feature extraction
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(lemmas)
x

pd.DataFrame.from_records([vectorizer.vocabulary_]).T.sort_values(0,ascending=True).head(10)

vectorizer.vocabulary_
```

```python
# bigram and trigram
vectorizer_ngram_range = CountVectorizer(analyzer='word',ngram_range=(1,3),max_features=200)    
bow_matrix_ngram = vectorizer_ngram_range.fit_transform(df)
bow_matrix_ngram
```

```python
print(vectorizer_ngram_range.get_feature_names())
print(bow_matrix_ngram.toarray())
```
